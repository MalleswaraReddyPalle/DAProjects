```python
import pandas as pd
#handling Ship Mode column values with 'Not Available' or 'unknown' by considering them as null values
df = pd.read_csv(r'orders.csv',encoding='utf-8',na_values=['Not Available','unknown'])
df.info()
```

```python
#standerdising column names 
#converting to lower case
df.columns=df.columns.str.lower()
#replacing space with underscore
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#Adding discount_price column using discount_percent,list_price columns
#df['discount_price'] = df['list_price']*df['discount_percent']*.01
#Adding sale_price column after discount using list_price-discount_price
#df['sale_price'] = df['list_price']-df['discount_price']
#Adding profit column using sale_price-cost_price
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
#changing order_date to appropriate date type
df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
df.dtypes
```
